```python
import pandas as pd
```

```python
file_path = '/workspaces/TIR104_g2/Ａ_raw_data/TWMovie2022-2025.csv'
dfJSON= pd.read_csv(file_path)

dfJSON_publisher = dfJSON[['MovieId', 'Publisher']]

print(dfJSON_publisher)
```

```python
new_df = dfJSON_publisher['Publisher'].str.split('##', expand = True)

print(new_df)
```

```python
dfJSON_publisher_new = pd.concat([dfJSON_publisher, new_df], axis=1)

dfJSON_publisher_new.drop(columns=['Publisher'], inplace= True)

dfJSON_publisher_new
```

```python
melted_df = pd.melt(dfJSON_publisher_new, id_vars = ['MovieId'])

melted_df.sort_values(by= 'value')
```

```python
melted_df.groupby('value').size().sort_values(ascending=False)
```

```python
melted_df[melted_df['value'].notna()].sort_values(by = "value")
```

```python
dfJSON_publisher_fin = melted_df[melted_df['value'].notna()].sort_values(by = "value")

dfJSON_publisher_fin.to_csv('/workspaces/TIR104_g2/Ａ_raw_data/TWMovie_publisher.csv', index = 0, encoding = 'utf-8-sig')
```
